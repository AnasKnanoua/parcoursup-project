```python
import pandas as pd
```

```python
df = pd.read_csv(r"/content/fr-esr-parcoursup.csv",sep = ";")
```

```python
df
```

```python
liste_de_filiere = list(df["Filière de formation"].unique())


df_moyenne_pro = pd.DataFrame(columns = ["filiere","moyenne"])
df_moyenne_techno = pd.DataFrame(columns = ["filiere","moyenne"])
df_moyenne_general = pd.DataFrame(columns = ["filiere","moyenne"])

def moyenne_filiere(dff,colonne):
    i = 0
    for filiere in liste_de_filiere:
        
        dfi = df[df["Filière de formation"]==filiere]
        a = dfi[[colonne,"Filière de formation"]]
        moyenne = a[colonne].mean()
        
        df2 = pd.DataFrame({"filiere":filiere,"moyenne":moyenne},index = [i])     
        
        dff = pd.concat([dff,df2])
        i+=1
    return dff
moyenne_filiere(df_moyenne_pro,r"% d’admis néo bacheliers professionnels")
moyenne_filiere(df_moyenne_general,r"% d’admis néo bacheliers généraux")
moyenne_filiere(df_moyenne_techno,r"% d’admis néo bacheliers technologiques")
```

```python
def moyenne_gen_pro_tech(filiere):
    df_prepa = df[df["Filière de formation"]==filiere]
    df_prepa = df_prepa[["Filière de formation","Effectif des admis néo bacheliers généraux","Effectif des admis néo bacheliers technologiques","Effectif des admis néo bacheliers professionnels",r"% d’admis néo bacheliers généraux",r"% d’admis néo bacheliers professionnels",r"% d’admis néo bacheliers technologiques"]]
    print(display(df_prepa))
    return df_prepa

for fil in liste_de_filiere:
    moyenne_gen_pro_tech(fil)
```

```python
liste_mentions = [r"% d’admis néo bacheliers sans information sur la mention au bac",r"% d’admis néo bacheliers sans mention au bac",r"% d’admis néo bacheliers avec mention Assez Bien au bac",r"% d’admis néo bacheliers avec mention Bien au bac",r"% d’admis néo bacheliers avec mention Très Bien au bac",r"% d’admis néo bacheliers avec mention Très Bien avec félicitations au bac"]

for mention in liste_mentions:
  for filiere in liste_de_filiere:
  
    df_mention = df[[mention,"Filière de formation"]].sort_values(by=mention,ascending = False)
    print(display(df_mention[df_mention["Filière de formation"]==filiere]))
```
